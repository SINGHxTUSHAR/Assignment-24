Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
sns.scatterplot(x = iris.petal_length , y = iris.petal_width)
```

```python
sns.displot(iris["sepal_length"])
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips, hue = 'smoker')
```

```python
sns.catplot(x = 'day' , y = 'total_bill' , data  = tips)
```

```python
sns.jointplot(x = tips.total_bill , y = tips.tip , color='#B2FF33')
```

```python
sns.regplot(x = tips.total_bill , y = tips.tip , data = tips , color="#2326F9")
```
